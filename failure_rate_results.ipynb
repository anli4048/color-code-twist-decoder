```python
import numpy as np
from scipy.optimize import minimize

fail_rates = np.vstack((
                       np.array([1-dc[(0,0,0)] for dc in results_15_thresh]),
                       np.array([1-dc[(0,0,0)] for dc in results_21_thresh]),
                       np.array([1-dc[(0,0,0)] for dc in results_27_thresh]),
                       np.array([1-dc[(0,0,0)] for dc in results_33_thresh])
                       ),
                    )

def pred_fail_rate (p,
                    d,
                    p_th,
                    alpha,
                    beta,
                    eta,
                    eps,
                    zeta,
                    ld,):
    return alpha * (p/p_th)**(beta*d + eta) * (1 + eps*(p/p_th)**(zeta*d**ld))

def pred_fail_rate_linear (p,
                    d,
                    p_th,
                    alpha,
                    beta,
                    eta,
                    eps,
                    zeta,
                    ld,):
    return alpha * (p/p_th)**(beta*d + eta)

p = np.linspace(0.08,0.10,20)
d = np.linspace(15,33,4)
p_mesh, d_mesh = np.meshgrid(p,d)

def least_squares_log (fail_rates,
                   p,
                   d,
                   params):
    return np.sum(np.sum((np.log(pred_fail_rate_linear(p,d,*params))-np.log(fail_rates))**2))

fit = minimize(lambda params: least_squares_log(fail_rates,p_mesh,d_mesh,params),[.5,.5,.5,.5,.5,.5,.5,],method='Nelder-Mead',tol=1e-6,options={'disp': True})

colors = ['black','red','blue','green']
plt.figure(figsize=(5,4), dpi=200)
for i in range(fail_rates.shape[0]):
    plt.scatter(p,fail_rates[i,:],1,color = colors[i], label = 'D={}'.format(int(d[i])))
    plt.errorbar(p,fail_rates[i,:],yerr=fail_rates[i,:]*0.006,fmt="None",capsize=2, ecolor = colors[i])
    # plt.plot(p,pred_fail_rate_linear(p_mesh,d_mesh,*fit.x)[i,:], color = colors[i], linestyle='dashed', label = 'Fitted logical error rate d={}'.format(d[i]))
plt.xlabel('Physical error rate')
plt.ylabel('String-like logical error rate')
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.savefig('plot')

print(fit.x)

print(p[14])
print((p[1]-p[0])*1.5)
                    
```
